# Literature Review

#### Text summarization is a Natural Language Processing (NLP) task that automatically generates a concise and meaningful summary from a long piece of text while preserving its main ideas. It helps users quickly understand the key information without reading the entire document. Text summarization is widely used in news websites, research paper analysis, business reports, legal documents, and educational applications.

#### Traditional text summarization methods relied on extractive techniques, where important sentences were selected directly from the original text. Although these methods are simple, they often produce summaries that are less coherent and may contain redundant information.

#### Recent advancements in Deep Learning and Transformer architectures have significantly improved text summarization. Transformer-based models understand the context and relationships between words more effectively than traditional approaches. Models such as BART, T5, and PEGASUS are widely used for abstractive text summarization because they can generate new sentences that accurately represent the meaning of the original text.

#### Among these models, BART (Bidirectional and Auto-Regressive Transformers) has shown excellent performance on benchmark datasets such as CNN/DailyMail and XSum. BART combines bidirectional encoding with autoregressive decoding, enabling it to generate fluent and high-quality summaries.

#### In this project, a pre-trained BART model from the Hugging Face Transformers library is used. Using a pre-trained model reduces training time while maintaining high summarization quality. The project also uses Streamlit to provide a simple and user-friendly web interface where users can enter text and receive an AI-generated summary.

#### Overall, Transformer-based text summarization provides an efficient solution for reducing reading time while preserving the essential information contained in long documents.

```python
## Import Regular Expression Library
import re
```

```python
## Create a Cleaning Function
def clean_text(text):
    text = re.sub(r"<.*?>", "", text)
    text = re.sub(r"[^a-zA-Z0-9\s]", "", text)
    text = text.lower()
    text = text.strip()
    return text
```

```python
## Clean the Articles
df["clean_article"] = df["article"].apply(clean_text)
```

```python
## Clean the Summaries
df["clean_summary"] = df["highlights"].apply(clean_text)
```

```python
from transformers import T5Tokenizer, T5ForConditionalGeneration
```

```python
## Load the Tokenizer
tokenizer = T5Tokenizer.from_pretrained("t5-small")
```

```python
## Model
model = T5ForConditionalGeneration.from_pretrained("t5-small")
```

```python
## First article
article = df["clean_article"][0]
print(article[:500])
```

```python
## T5 input create
input_text = "summarize: " + article
```

```python
## Tokenize
inputs = tokenizer(
    input_text,
    return_tensors="pt",
    max_length=512,
    truncation=True
)
```

```python
## Generate Summary
summary_ids = model.generate(
    inputs["input_ids"],
    max_length=60,
    min_length=20,
    num_beams=4,
    early_stopping=True
)
```

```python
## Decode Summary
summary = tokenizer.decode(
    summary_ids[0],
    skip_special_tokens=True
)

print("Generated Summary:")
print(summary)
```
